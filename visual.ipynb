```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

instruments = 100
all_instruments = [i for i in range(instruments)]
df = pd.read_csv("./prices.txt", delim_whitespace=True, names=all_instruments)
arr = df[all_instruments]
```

```python
for pos in all_instruments:
    plt.plot(arr[pos])
    plt.title(f"Instrument {pos}")
    plt.show()
```

```python
for inst in arr:
    plot_acf(df[inst], title = f"Original autocorrelation for instrument {inst}")
    diff1 = df[inst].diff()
    diff1.iloc[0] = 0
    plot_acf(diff1, title=f"1st Order correlation for instrument {inst}")
    diff2 = diff1.diff()
    diff2.iloc[0] = 0
    plot_acf(diff2, title=f"2nd Order correlation for instrument {inst}")
```
